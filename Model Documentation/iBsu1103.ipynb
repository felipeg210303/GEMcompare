### Henry et al., 2009

Ahora, el tercer modelo utilizado es iBsu1103 que contó con los siguientes cambios frente a iYO844 de Oh et al., 2008:
* Incluyó datos termodinámicos para determinar la direccionalidad de las reacciones
* Desagrupó reacciones del anterior modelo, también conocidas como "lumped reactions"
* Agrega más de 600 nuevas reacciones

De igual forma a como con los demás modelos, se importan las librerías:

```python
import cobra
import matplotlib.pyplot as plt
import numpy as np

```

Se carga el modelo en lenguaje SBML

```python
model = cobra.io.read_sbml_model("C:/Users/felip/Desktop/python/TESIS/iBsu1103.xml")

solution = model.optimize()

# Mostrar el valor objetivo
print(f'Valor objetivo: {solution.objective_value}')
```

```python
# Obtener los flujos de las reacciones
fluxes = solution.fluxes

# Graficar los flujos de reacción
reactions = list(fluxes.index)
values = list(fluxes.values)

plt.figure(figsize=(10, 6))
plt.barh(reactions, values, color="skyblue")

plt.axvline(0, color="grey", lw=0.8)  # Línea vertical en x=0
plt.title("Flujos de Reacción del Modelo Metabólico")
plt.xlabel("Flujo")
plt.ylabel("Reacciones")

plt.grid(axis='x')
plt.show()
```

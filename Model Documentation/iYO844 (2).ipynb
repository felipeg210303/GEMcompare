# iYO844 (2007)

Authors: You-Kwan Oh, Bernhard O. Palsson, Sung M. Park, Christophe H. Schilling, and Radhakrishnan Mahadevan<br>
Publication date: June 15, 2007<br>
Summary: This was the first genome-scale metabolic model of *B. subtilis*. The authors used the first availabel annotation data to construct the model and then used the high-throughput phenotyping data to validate the it. They added 48 missing reactions needed for growth from this annotation through gap-filling. With the high-throughput phenotyping data, they were able to identify 75 more missing reactions needed for more accurate predictions The model predicted accurately the growth in 200 out of 271 substrates (74%) but since the full annotation of the genome was incomplete it lacked to represent key metabolic capabilities of the microorganism.<br>
Scientific paper:  [Genome-scale Reconstruction of Metabolic Network in *Bacillus subtilis* Based on High-throughput Phenotyping and Gene Essentiality Data](https://doi.org/10.1074/jbc.M703759200) <br>
Source of the model:  [BiGG Models](http://bigg.ucsd.edu/models/iYO844) <br>

First of all, we add the libraries needed to use the model:

```python
import cobra
import pandas as pd
import matplotlib.pyplot as plt
```

Now, we load the model (which we should have arleady downloaded) in a COBRA object called model.


```python
model=cobra.io.read_sbml_model("iYO844_bigg.xml") # path of the model
```

Then, we can see the properties of the model in a table by just calling the model object. This model has 987 metabolites, 1245 reactions and 844 genes.

```python
model
```

Also, we can see a summary of model which is a list of the uptake and secretion reactions with their fluxes and the value of the objective function that by default is biomass.


```python
model.summary()
```
